```python
data1 = data[["AuthDate", "PrdId", "SchId", "BaseQty", "Loss","Profit","PrdGrossAmount"]]
data1.head()
```

```python
D1 = data1[data.AuthDate == '2013-02-18']
```

```python
D1[D1.Profit != 0]
```

```python
D2 =D1.groupby(['AuthDate','PrdId','SchId','Profit',"PrdGrossAmount"])['BaseQty'].max()
D2=pd.DataFrame(D2)
D2.head()
```

```python
D2.reset_index(inplace=True)
D2.head()
```

```python
D2 = D2[D2.Profit != 0]
D2 = D2.groupby(["AuthDate",'PrdId','SchId'])['BaseQty','Profit',"PrdGrossAmount"].sum()
D2.head()
```

```python
D2 = D2.sort_values(by=['BaseQty'], ascending=False)
D2.reset_index(inplace=True)
D2.head()
```

```python
# Base Qty what is profit of same Product
from pulp import *
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
prob = pulp.LpProblem("maximize_production", pulp.LpMaximize)
```

```python
decision_variable = []
for rownum,row in D2.iterrows():
    variable = str('X'+ str(rownum))
    variable = pulp.LpVariable(str(variable), lowBound= 0, upBound= 1, cat="Integer")
    decision_variable.append(variable)
```

```python
print ("Total number of decision_variables: " + str(len(decision_variable)))
print ("Array with Decision Variables:" + str(decision_variable))
```

```python
total_Profit = ""

for rownum, row in D2.iterrows():
    for i, shedule in enumerate(decision_variable):
        if rownum == i:
            formula = row['Profit']*shedule
            total_Profit += formula
            
prob += total_Profit

print ("Optimization function: " + str(total_Profit))
```

```python
aval_BaseQty  = 150
total_BaseQty = ""
for rownum, row in D2.iterrows():
	for i, schedule in enumerate(decision_variable):
		if rownum == i:
			formula = row['BaseQty']*schedule
			total_BaseQty += formula

prob += (total_BaseQty == aval_BaseQty)
```

```python
prob
```

```python
print(prob)
prob.writeLP("GoingOnVacation.lp" )
```

```python
optimization_result = prob.solve()

#assert optimization_result == pulp.LpStatusOptimal
print("Status:", LpStatus[prob.status])
print("Optimal Solution to the problem: ", value(prob.objective))
print ("Individual decision_variables: ")
for v in prob.variables():
	print(v.name, "=", v.varValue)
```

```python
variable_name = []
variable_value = []

for v in prob.variables():
	variable_name.append(v.name)
	variable_value.append(v.varValue)

D3 = pd.DataFrame({'variable': variable_name, 'value': variable_value})
```

```python
for rownum, row in D3.iterrows():
	value = re.findall(r'(\d+)', row['variable'])
	D3.loc[rownum, 'variable'] = int(value[0])

D3 = D3.sort_index(by='variable')

#append results
for rownum, row in D2.iterrows():
	for results_rownum, results_row in D3.iterrows():
		if rownum == results_row['variable']:
			D2.loc[rownum, 'decision'] = results_row['value']
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import pandas as pd
import dash_table
from dash.dependencies import Input, Output
import plotly.graph_objs as go

external_css = ["https://cdnjs.cloudflare.com/ajax/libs/skeleton/2.0.4/skeleton.min.css",
                "https://fonts.googleapis.com/css?family=Raleway:400,400i,700,700i",
                "https://fonts.googleapis.com/css?family=Product+Sans:400,400i,700,700i"]


app = dash.Dash(external_stylesheets=external_css)

df = pd.read_csv('main_final.csv')

def generate_table(dataframe, max_rows=10):
    return html.Table(
        # Header
        [html.Tr([html.Th(col) for col in dataframe.columns])] +

        # Body
        [html.Tr([
            html.Td(dataframe.iloc[i][col]) for col in dataframe.columns
        ]) for i in range(min(len(dataframe), max_rows))]
    )

PrdId_options = []
for PrdId in df['PrdId'].unique():
    PrdId_options.append({'label':str(PrdId),'value':PrdId})

# Create a Dash layout
app.layout = html.Div([
    html.Div(
        html.H1('My Dashboard')
    ),
    dcc.Tabs(id="tabs", value='Tab1', children=[
      dcc.Tab(label='Tab 0', id='tab0', value= 'Tab0', children=[
            html.H1(html.A(
                'Botree Data',
                style={'text-decoration': 'none', 'color': 'inherit'}
            )),
            html.H2(html.A(
                html.Img(src="https://images.ctfassets.net/k49d63tr8kcn/7Ay4DJqdtSoisK8iGMqqwk/8a2ff4a82f22d7764e2ea2152a56bc94/plotly.png"),
                href='https://plot.ly/products/dash/'
            )),
           dcc.Dropdown(
                id='dropdown1',
                options=[{'label' : i, 'value' : i} for i in df.PrdId.unique()],
                multi=True,
                placeholder = "Filter by PrdId"
            ),
          html.Div(id='tabs-content'),
          ]),
        
        dcc.Tab(label='Tab 1', id='tab1', value='Tab1', children =[
            dcc.Dropdown(
                id='dropdown',
                options=[{'label' : i, 'value' : i} for i in df.PrdId.unique()],
                multi=True,
                placeholder = "Filter by PrdId"
            ),
            dcc.Graph(id='graph-1'),
        ]),

        dcc.Tab(label='Tab 2', id='tab2', value= 'Tab2', children=[
            dcc.Dropdown(
                id='dropdown2',
                options=[{'label' : i, 'value' : i} for i in df.PrdId.unique()],
                multi=True,
                placeholder = "Filter by PrdId"
            ),
            dcc.Graph(id='graph-2'),
        ]),
        
        dcc.Tab(label = 'Tab 3', id= 'tab 2', value = 'Tab3', children = [
            dcc.Dropdown(
            id= 'dropdown3',
            options = [{'label' : i, 'value' : i} for i in df.PrdId.unique()],
            multi = True,
            placeholder = "Filter by PrdId"
            ),
            dcc.Graph(id = 'graph-3'),
            dcc.Graph(id = 'graph-4'),

        ])
    ])
])


@app.callback(
    dash.dependencies.Output('tabs-content', 'children'),
    [dash.dependencies.Input('dropdown1', 'value')])

def display_table(dropdown_value):
    if dropdown_value is None:
        return generate_table(df)
    else:
        dff = pd.DataFrame()
        for val in dropdown_value:
            temp = df[df['PrdId'] ==val]
            dff = pd.concat([dff,temp], axis = 0)
        return generate_table(dff)



@app.callback(Output('graph-1', 'figure'),
              [Input('dropdown', 'value')])
def update_figure_1(selected_PrdId):
    print('selected product id is {}'.format(selected_PrdId))
    filtered_df = df[df['PrdId'] == selected_PrdId]
    traces = []
    for PrdId_name in filtered_df['PrdId'].unique():
        df_by_continent = filtered_df[filtered_df['PrdId'] == PrdId_name]
        traces.append(go.Scatter(
            x=df_by_continent['SchId'],
            y=df_by_continent['profit'],
            text=df_by_continent['BaseQty'],
            mode='markers',
            opacity=0.7,
            #marker={'size': 15,'color' :'rgba(255, 182, 193, .9)'},
            marker={'size': 15,'line': {'width': 0.5, 'color': 'black'},'color': 'rgba(255, 182, 193, .9)'},
            name='continent_name',
        ))

    return {
        'data': traces,
        'layout': go.Layout(
                xaxis={'type': 'log', 'title': 'ProductId'},
                yaxis={'title': 'Profit'},
                margin={'l': 40, 'b': 40, 't': 10, 'r': 10},
                legend={'x': 0, 'y': 1},
                hovermode='closest'
            )
        #go.Layout(
           # xaxis={'type': 'log', 'title': 'ProductId'},
           # yaxis={'title': 'Profit'},
           # hovermode='closest'
        #)
    }


@app.callback(Output('graph-2', 'figure'),
              [Input('dropdown2', 'value')])
def update_figure_1(selected_PrdId):
    print('selected product id is {}'.format(selected_PrdId))
    filtered_df = df[df['PrdId'] == selected_PrdId]
    traces = []
    for PrdId_name in filtered_df['PrdId'].unique():
        df_by_continent = filtered_df[filtered_df['PrdId'] == PrdId_name]
        traces.append(go.Bar(
            x=df_by_continent['SchId'],
            y=df_by_continent['PrdGrossAmount'],
            #text=df_by_continent['country'],
             ))

    return {
        'data': traces,
        'layout': go.Layout(
            xaxis={'type': 'log', 'title': 'Product ID'},
            yaxis={'title': 'PrdGrossAmount'},
            hovermode='closest'
        )
    }


@app.callback(Output('graph-3', 'figure'),
              [Input('dropdown3', 'value')])
def update_figure_1(selected_PrdId):
    print('selected product id is {}'.format(selected_PrdId))
    filtered_df = df[df['PrdId'] == selected_PrdId]
    traces = []
    for PrdId_name in filtered_df['PrdId'].unique():
        df_by_continent = filtered_df[filtered_df['PrdId'] == PrdId_name]
        traces.append(go.Bar(
            x=df_by_continent['month'],
            y=df_by_continent['PrdGrossAmount'],
            #text=df_by_continent['country'],
             ))

    return {
        'data': traces,
        'layout': go.Layout(
            xaxis={'type': 'log', 'title': 'months'},
            yaxis={'title': 'PrdGrossAmount'},
            hovermode='closest'
        )
    }


@app.callback(Output('graph-4', 'figure'),
              [Input('dropdown3', 'value')])
def update_figure_1(selected_PrdId):
    print('selected product id is {}'.format(selected_PrdId))
    filtered_df = df[df['PrdId'] == selected_PrdId]
    traces = []
    for PrdId_name in filtered_df['PrdId'].unique():
        df_by_continent = filtered_df[filtered_df['PrdId'] == PrdId_name]
        traces.append(go.Bar(
            x=df_by_continent['day'],
            y=df_by_continent['PrdGrossAmount'],
            #text=df_by_continent['country'],
             ))

    return {
        'data': traces,
        'layout': go.Layout(
            xaxis={'type': 'log', 'title': 'Day'},
            yaxis={'title': 'PrdGrossAmount'},
            hovermode='closest'
        )
    }



if __name__ == '__main__':
    app.run_server(port = 2345,debug=False)
```
